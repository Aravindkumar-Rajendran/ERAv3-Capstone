```python
# load the pdf file and extract the content using pypdf2
import PyPDF2
def extract_pdf_content(pdf_path):
    content = ""
    try:
        with open(pdf_path, "rb") as file:
            reader = PyPDF2.PdfReader(file)
            for page in reader.pages:
                content += page.extract_text() + "\n"
    except Exception as e:
        print(f"An error occurred while reading the PDF: {e}")
    return content.strip()

pdf_path = "../../test_chapter.pdf"  
content = extract_pdf_content(pdf_path)
if content:
    print("Extracted Content:")
    print(content)
else:
    print("No content extracted or the PDF might be empty.")
```

```python
# unstructured pdf extraction using langchain
from langchain_community.document_loaders import PyPDFLoader

def extract_pdf_content_langchain(pdf_path):
    try:
        loader = PyPDFLoader(pdf_path)
        documents = loader.load()
        content = "\n".join([doc.page_content for doc in documents])
    except Exception as e:
        print(f"An error occurred while reading the PDF with Langchain: {e}")
        content = ""
    return content.strip()

pdf_path_langchain = "../../test_chapter.pdf"
content_langchain = extract_pdf_content_langchain(pdf_path_langchain)
if content_langchain:
    print("Extracted Content using Langchain:")
    print(content_langchain)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=0)
texts = text_splitter.split_text(content_langchain)
print(f"Number of chunks: {len(texts)}")
```

```python
# unstructured pdf extraction using langchain
from langchain.document_loaders.pdf import UnstructuredPDFLoader

def extract_pdf_content_unstructured(pdf_path):
    try:
        loader = UnstructuredPDFLoader(pdf_path)
        documents = loader.load()
        content = "\n".join([doc.page_content for doc in documents])
    except Exception as e:
        print(f"An error occurred while reading the PDF with Unstructured: {e}")
        content = ""
    return content.strip()

pdf_path_unstructured = "test_chapter.pdf"
content_unstructured = extract_pdf_content_unstructured(pdf_path_unstructured)
if content_unstructured:
    print("Extracted Content using Unstructured:")
    print(content_unstructured)
```

```python
from utils.gemini_client import GeminiClient

# Initialize the Gemini client
client = GeminiClient()
```

```python
prompt = """Your Role: You are an expert analyst and editor. Your task is to deconstruct and simplify complex text into a clear, concise, and easily digestible format.

Objective: Analyze the provided text below. Your goal is to identify the main topics, clean up the content, and present it in a structured summary.

Instructions:

Analyze the following text and perform these steps:

1. Segment the Text: Read through the entire text and divide it into logical, coherent chunks. The number of chunks should be determined by the natural shifts in topic. Keep the chunks and topics in high level and max of 4.
2. For Each Chunk, Process as Follows:
    - Identify a Core Topic: Create a clear and descriptive heading of 3-5 words that accurately reflects the main theme of the chunk.
    - Clean and Refine:
        - Remove any repeated sentences or redundant phrases.
        - Eliminate non-essential special characters (e.g., emojis, artifacts), but retain standard punctuation (periods, commas, etc.) needed for readability.
        - Normalize whitespace by trimming leading/trailing spaces from the chunk and ensuring only a single space between words.
3. Clean the Essence: Clean the content by removing repeated words and removing characters. Preserve most the content only removing the non-essential characters.


Output Format:
Present the final output strictly in the following JSON format. Do not add any introductory or concluding remarks.

JSON

[
  {
    "topic": "[Insert the 3-5 word heading here]",
    "content": "[Insert the summarized and cleaned content for the first chunk here]"
  },
  {
    "topic": "[Insert the 3-5 word heading here]",
    "content": "[Insert the summarized and cleaned content for the second chunk here]"
  },
  // Repeat this structure for every chunk you identify
]
Text to Analyze:
""" + content_langchain

# Call the Gemini model to get the summary
chunks = client.generate(prompt)

print("Chunking and Topic generation using Gemini:")
print(chunks)
```
